```python
df = df.rename(columns={
    'Percentage of adult employment that is part-time_women': 'parttime_women',
    'Percentage of adult employment that is part-time_men': 'parttime_men',
    "Women's share of part-time employment": 'women_share'
})
# --- データに Continent 列を追加 ---
df['Continent'] = df['Country or area'].map(country_to_continent)

# --- うまくマッチしなかった国（要修正）を確認 ---
unmatched = df[df['Continent'].isna()]['Country or area'].unique()
print("未マッチの国:", unmatched)
```

```python

# convert numeric
for col in ['parttime_women', 'parttime_men', 'women_share']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

summary = df[['parttime_women','parttime_men','women_share']].describe()
missing = df.isnull().sum()

summary, missing
```

```python
# Histogram: women part-time
plt.figure()
plt.hist(df['parttime_women'].dropna())
plt.xlabel('Part-time employment (women, %)')
plt.ylabel('Count')
plt.title('Distribution of Women Part-time Employment')
plt.show()
```

```python
# Histogram: men part-time
plt.figure()
plt.hist(df['parttime_men'].dropna())
plt.xlabel('Part-time employment (men, %)')
plt.ylabel('Count')
plt.title('Distribution of Men Part-time Employment')
plt.show()
```

```python
plt.figure()
plt.scatter(df['parttime_men'], df['parttime_women'])
plt.xlabel('Men Part-time (%)')
plt.ylabel('Women Part-time (%)')
plt.title('Women vs Men Part-time Employment')
plt.show()
```

```python
# Boxplot
plt.figure()
plt.boxplot([df['parttime_women'].dropna(), df['parttime_men'].dropna()],
            labels=['Women', 'Men'])
plt.ylabel('Part-time employment (%)')
plt.title('Boxplot of Part-time Employment (Women vs Men)')
plt.show()
```

```python
# Q1 calculations

df_valid = df.dropna(subset=['parttime_women','parttime_men'])
female_more = (df_valid['parttime_women'] > df_valid['parttime_men']).sum()
male_more = (df_valid['parttime_men'] > df_valid['parttime_women']).sum()
female_more, male_more
```

```python
# top 10 highest women_share
top10 = df.dropna(subset=['women_share']).sort_values('women_share', ascending=False).head(20)

top10
```

```python
continents = df_valid['Continent'].unique()
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown']
cont_to_color = {c: colors[i % len(colors)] for i, c in enumerate(continents)}

plt.figure()
for c in continents:
    sub = df_valid[df_valid['Continent'] == c]
    plt.scatter(sub['parttime_men'], sub['parttime_women'],
                label=c, color=cont_to_color[c])
plt.plot([0, df_valid['parttime_men'].max()],
         [0, df_valid['parttime_men'].max()],
         linestyle='--')
plt.xlabel('Men Part-time (%)')
plt.ylabel('Women Part-time (%)')
plt.title('Q1: Women vs Men Part-time by Continent')
plt.legend()
plt.show()
```

```python
# ===== 大陸別平均 =====
continent_means = df_valid.groupby('Continent')[['parttime_women','parttime_men','women_share']].mean()

# 女性パート率平均
plt.figure()
continent_means['parttime_women'].plot(kind='bar')
plt.ylabel('Average Women Part-time (%)')
plt.title('Average Women Part-time by Continent')
plt.show()

# 男性パート率平均
plt.figure()
continent_means['parttime_men'].plot(kind='bar')
plt.ylabel('Average Men Part-time (%)')
plt.title('Average Men Part-time by Continent')
plt.show()

# パートに占める女性シェア平均
plt.figure()
continent_means['women_share'].plot(kind='bar')
plt.ylabel('Average Women Share in Part-time (%)')
plt.title('Average Women Share in Part-time by Continent')
plt.show()
```

```python
# 大陸 → 色 の対応
continents = sorted(df_valid['Continent'].unique())
base_colors = ['tab:blue', 'tab:orange', 'tab:green',
               'tab:red', 'tab:purple', 'tab:brown']
cont_to_color = {c: base_colors[i % len(base_colors)]
                 for i, c in enumerate(continents)}

# 凡例用パッチ
legend_handles = [
    mpatches.Patch(color=cont_to_color[c], label=c) for c in continents
]
```

```python
# Top5 women
plt.figure()
colors_top_w = [cont_to_color[c] for c in top_women['Continent']]
plt.bar(top_women['Country or area'], top_women['parttime_women'],
        color=colors_top_w)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Women Part-time (%)')
plt.title('Top 10 Countries (Women Part-time) by Continent')
plt.legend(handles=legend_handles, title='Continent')
plt.tight_layout()
plt.show()
```

```python
plt.figure()
colors_bottom_w = [cont_to_color[c] for c in bottom_women['Continent']]
plt.bar(bottom_women['Country or area'], bottom_women['parttime_women'],
        color=colors_bottom_w)
plt.xticks(rotation=45, ha='right')
plt.ylabel('Women Part-time (%)')
plt.title('Bottom 10 Countries (Women Part-time) by Continent')
plt.legend(handles=legend_handles, title='Continent')
plt.tight_layout()
plt.show()
```
